```python
# Dropando colunas que não fazem parte da análise
df = df.drop(columns=['product_link', 'review_id', 'review_title', 'review_content', 'img_link', 'about_product'])
df_dup = df.drop_duplicates() # Removendo linhas duplicadas, pois não representam transações separadas
```

```python
# Tirando virgulas e simbolos, para realizar a conversão
def replacing_values(df):
    # Remover caracteres não numéricos de cada coluna
    df['discounted_price'] = df['discounted_price'].astype(str).str.replace(r'\D', '', regex=True)
    df['actual_price'] = df['actual_price'].astype(str).str.replace(r'\D', '', regex=True)
    df['rating_count'] = df['rating_count'].astype(str).str.replace(r'\D', '', regex=True)
    
    # Converter as colunas de volta para numérico
    df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
    df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
    df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
    return df

df = replacing_values(df)

# Transformando Rupia Indiana para Dólar, atualmnte a cotação está a seguinte : $1,00 == ₹84,07 or ₹1,00 == $0.012
df['actual_price'] = df['actual_price'].apply(lambda x: round(int(x) / 83.33, 2))
df['discounted_price'] = df['discounted_price'].apply(lambda x: round(int(x) / 83.33, 2))

df.head(4)
```
